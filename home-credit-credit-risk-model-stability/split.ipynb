```python
def generate_create_table_sql(table_name, headers):
    columns = ",\n    ".join([f"[{col}] NVARCHAR(MAX)" for col in headers])
    sql = f"CREATE TABLE {table_name} (\n    {columns}\n);"
    return sql
```

```python
import sqlalchemy

engine = sqlalchemy.create_engine(
    "mssql+pyodbc://username:password@localhost/YourDatabase?driver=ODBC+Driver+17+for+SQL+Server"
)

def create_table(engine, table_name, headers):
    sql = generate_create_table_sql(table_name, headers)
    with engine.connect() as conn:
        conn.execute(sqlalchemy.text(sql))
```

```python
def bulk_insert(engine, table_name, csv_file_path):
    cmd = f"""
    BULK INSERT {table_name}
    FROM '{csv_file_path}'
    WITH (
        FORMAT='CSV',
        FIRSTROW=2,
        FIELDTERMINATOR=',',
        ROWTERMINATOR='\\n'
    )
    """
    with engine.connect() as conn:
        conn.execute(sqlalchemy.text(cmd))
```

```python
for file in csv_files:
    headers = get_headers(file)
    table_name = "dbo." + file.split(".")[0]  # e.g. data1  dbo.data1
    create_table(engine, table_name, headers)
    bulk_insert(engine, table_name, file)
```
