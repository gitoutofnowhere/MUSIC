```python
import re

def normalize_col(col):
    col = col.strip().lower()
    col = re.sub(r"[^a-z0-9_]", "_", col)
    if col[0].isdigit():
        col = "_" + col
    return col
```

```python
def generate_create_table_sql(table_name, columns):
    cols_sql = ",\n".join(
        f"{col} TEXT" for col in columns
    )
    return f"""
    DROP TABLE IF EXISTS {table_name};
    CREATE TABLE {table_name} (
        {cols_sql}
    );
    """
```

```python
from sqlalchemy import create_engine, text

engine = create_engine(
    "postgresql+psycopg2://postgres:123456@localhost:5432/postgres"
)

def create_table_and_copy(csv_path, table_name):
    raw_header = read_csv_header(csv_path)
    columns = [normalize_col(c) for c in raw_header]

    create_sql = generate_create_table_sql(table_name, columns)

    # FIX IS HERE ðŸ‘‡
    with engine.begin() as conn:
        conn.execute(text(create_sql))

    with open(csv_path, "r", encoding="utf-8") as f:
        raw_conn = engine.raw_connection()
        cursor = raw_conn.cursor()

        col_list = ", ".join(columns)

        copy_sql = f"""
        COPY {table_name} ({col_list})
        FROM STDIN
        WITH (FORMAT CSV, HEADER TRUE)
        """

        cursor.copy_expert(copy_sql, f)
        raw_conn.commit()

        cursor.close()
        raw_conn.close()

```

```python
import os

train = "C:\\Users\\ASUS\\Music\\home-credit-credit-risk-model-stability\\train"

for file in os.listdir(train):
    if file.endswith(".csv"):
        table_name = file.replace(".csv", "").lower()
        csv_path = os.path.join(train, file)

        print(f"Processing {file} â†’ {table_name}")
        create_table_and_copy(csv_path, table_name)
```
